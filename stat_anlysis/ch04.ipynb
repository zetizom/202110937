```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

%precision 3
%matplotlib inline
```

```python
df = pd.read_csv('../stat_anlysis/data/ch4_scores400.csv')
scores = np.array(df['score'])
scores[:10]

# 빨간색 글자들은 전부 시험에 나올 수 있음 
# 모집단, 표본, 표본 추출, 표본 크기, 표본 통계량, 모수
# 추정량, 추정값 중요
```

```python
# 비복원 ,복원 추출 중요 외우기 
np.random.choice([1,2,3], 3) # 표본에서 3개 뽑기
```

```python
np.random.choice([1, 2, 3], 3, replace = False) #표본에서 3개뽑기 단, 중복 x
```

```python
np.random.seed(0)
np.random.choice([1, 2, 3], 3)
# 시드를 0으로 하고 추출
# 같은 시드값은 항상 동일하게 나옴 
```

```python
np.random.seed(0)
sample = np.random.choice(scores, 20) 
sample.mean()

# 20번 렌덤 뽑기한 걸 평균화(표본 평균), 실제 편균 69와 얼마 차이가 안남 
```

```python
# 중요 표본추출은 항상 같은 값이 나오는 것은 아니다

for i in range(5):
    sample = np.random.choice(scores, 20)
    print(f'{i+1}번째 무작위 추출로 얻은 표본평균', sample.mean())
```

```python
# 확률, 확률 모형, 확률변수(중요), 시행, 실현값, 사건(중요), 근원사건(매우중요), 상호배반(중요)  

dice = [1, 2, 3, 4, 5, 6]
prob = [1/21, 2/21, 3/21, 4/21, 5/21, 6/21]
np.random.choice(dice,p = prob)
```

```python
num_trial = 100
sample = np.random.choice(dice, num_trial, p=prob)
sample 
# 확률이 다른 주사위를 사용한 경우
```

```python
freq, _= np.histogram(sample, bins=6, range=(1, 7))
pd.DataFrame({' frequency ': freq,
               ' relative frequency ' :freq / num_trial},
              index = pd.Index(np.arange(1, 7), name= 'dice'))
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.hist(sample, bins=6, range=(1, 7), density=True, rwidth=0.8)
ax.hlines(prob, np.arange(1, 7), np.arange(2, 8), colors = 'gray' )
ax.set_xticks(np.linspace(1.5, 6.5, 6))
ax.set_xticklabels(np.arange(1, 7))
ax.set_xlabel('dice')
ax.set_ylabel('relative frequency')
plt.show()
```

```python
num_trial = 10000
sample = np.random.choice(dice, num_trial, p=prob)


fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.hist(sample, bins=6, range=(1, 7), density=True, rwidth=0.8)
ax.hlines(prob, np.arange(1, 7), np.arange(2, 8), colors = 'gray' )
ax.set_xticks(np.linspace(1.5, 6.5, 6))
ax.set_xticklabels(np.arange(1, 7))
ax.set_xlabel('dice')
ax.set_ylabel('relative frequency')
plt.show()
```

```python
sample_means = [np.random.choice(scores, 20).mean()
               for _ in range(10000)]

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.hist(sample_means, bins=100, range=(0, 100), density=True)
ax.vlines(np.mean(scores), 0, 1, 'gray')
ax.set_xlim(50, 90)
ax.set_ylim(0, 0.13)
ax.set_xlabel('score')
ax.set_ylabel('relative frequency')
plt.show()

# scores를 10000번 반복한 값이 어디에 많이 몰려있는지 확인하는 방법으로 즉, 표본평균으로 실제 평균(모평균)을 추측할 수 있다.
```
