```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

%precision 3
%matplotlib inline
```

```python
df = pd.read_csv('../stat_anlysis/data/ch4_scores400.csv')
scores = np.array(df['score'])
scores[:10]
```

```python
np.random.choice([1,2,3], 3) # 표본에서 3개 뽑기
```

```python
np.random.choice([1, 2, 3], 3, replace = False) #표본에서 3개뽑기 단, 중복 x
```

```python
np.random.seed(0)
np.random.choice([1, 2, 3], 3)
# 시드를 0으로 하고 추출
```

```python
np.random.seed(0)
sample = np.random.choice(scores, 20) 
sample.mean()

# 20번 렌덤 뽑기한 걸 평균화(표본 평균), 실제 편균 69와 얼마 차이가 안남 
```

```python
for i in range(5):
    sample = np.random.choice(scores, 20)
    print(f'{i+1}번째 무작위 추출로 얻은 표본평균', sample.mean())
```

```python
dice = [1, 2, 3, 4, 5, 6]
prob = [1/21, 2/21, 3/21, 4/21, 5/21, 6/21]
```

```python
num_trial = 100
sample = np.random.choice(dice, num_trial, p=prob)
sample 
# 확률이 다른 주사위를 사용한 경우
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.hist(scores, bins=100, range=(0, 100), density=True)
ax.set_xlim(20, 100)
ax.set_ylim(0, 0.042)
ax.set_xlabel('score')
ax.set_ylabel('relative frequency')
plt.show()

# scores array를 다 가져온 후 간격을 1로 한 히스토그램을 그림
```

```python
np.random.choice(scores) # score에서 랜덤 출력
```

```python
sample = np.random.choice(scores, 10000)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.hist(sample, bins=100, range=(0, 100), density=True)
ax.set_xlim(20, 100)
ax.set_ylim(0, 0.042)
ax.set_xlabel('score')
ax.set_ylabel('relative frequency')
plt.show()

# scores에서 10000개 뽑은 값들을 히스토그램으로 넣은 경우 실제 분포와 비슷하게 나온다. 표본이 커지면 커질 수록 더 유사
```

```python
sample_means = [np.random.choice(scores, 20).mean()
               for _ in range(10000)]

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.hist(sample_means, bins=100, range=(0, 100), density=True)
ax.vlines(np.mean(scores), 0, 1, 'gray')
ax.set_xlim(50, 90)
ax.set_ylim(0, 0.13)
ax.set_xlabel('score')
ax.set_ylabel('relative frequency')
plt.show()

# scores를 10000번 반복한 값이 어디에 많이 몰려있는지 확인하는 방법으로 즉, 표본평균으로 실제 평균(모평균)을 추측할 수 있다.
```
